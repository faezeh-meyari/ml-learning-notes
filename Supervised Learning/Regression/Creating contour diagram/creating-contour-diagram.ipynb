```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Required for 3D plotting
```

```python
# Define the range of x and y values
x= np.arange(-10,10,0.5)
y= np.arange(-10,10,0.5)
print(x.shape)
print(x)
```

```python
# Create a 2D grid of x and y values
X,Y = np.meshgrid(x,y)
print(X.shape)
print(X)
```

```python
# Define the Z values based on a mathematical function
Z=X**2+Y**2 # This is a paraboloid surface
```

```python
# Create a figure and a 3D subplot
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d') # 111 means 1 row, 1 column, 1 plot
# Plot the 3D surface
surf = ax.plot_surface(X,Y,Z , cmap=plt.cm.rainbow)
# Add filled contour plot projected onto the bottom plane (z=0)
cset = ax.contourf(X,Y,Z, zdir='z', offset=-0, cmap=plt.cm.rainbow)
# cset = ax.contour(X,Y,Z, zdir='z', offset=-0, cmap=plt.cm.rainbow)

# Add a color bar for reference
fig.colorbar(surf,shrink=0.5, aspect=5)
# Display the plot
plt.show()
```
